```python
import asyncio
from playwright.async_api import async_playwright

async def run_search():
    async with async_playwright() as p:
        # 启动浏览器 (headless=False 可以让你看到操作过程)
        browser = await p.chromium.launch(headless=False)
        context = await browser.new_context(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
        )
        page = await context.new_page()

        # 1. 访问目标 URL
        url = "https://standards.cencenelec.eu/ords/f?p=CEN:105"
        print(f"正在访问: {url}")
        await page.goto(url, wait_until="networkidle")

        # 2. 定位搜索框
        # 该页面搜索框通常是一个 input[type="text"]，我们使用更稳健的选择器
        search_selector = "input[type='text']" 
        await page.wait_for_selector(search_selector)
        
        # 3. 输入关键字并回车
        keyword = "62443"  # 替换为你想要搜索的关键字
        print(f"正在搜索关键字: {keyword}")
        await page.fill(search_selector, keyword)
        await page.keyboard.press("Enter")

        # 4. 等待搜索结果加载
        # 注意：APEX 页面搜索后通常会刷新局部或整个页面
        # 我们等待包含结果的表格出现，例如含有 'Standard reference' 文本的元素
        try:
            # 等待网络请求完成或特定结果列表元素出现
            await page.wait_for_load_state("networkidle")
            
            # 5. 提取搜索结果 (以提取结果表格的文本为例)
            # 这里需要根据搜索后的页面实际结构调整选择器
            results = await page.locator("table.t-Report-report").all_inner_texts()
            
            for index, text in enumerate(results):
                print(f"结果块 {index + 1}:")
                print(text)
                
        except Exception as e:
            print(f"获取结果时出错: {e}")

        # 保持浏览器开启一段时间以便查看结果，然后关闭
        await asyncio.sleep(5)
        await browser.close()

# 在 Jupyter 中直接 await 协程，避免 asyncio.run 冲突
await run_search()
```
